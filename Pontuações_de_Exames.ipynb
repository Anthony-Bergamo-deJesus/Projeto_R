## Introdução

O conjunto de dados aborda informações sobre trabalho, família e renda. Seu objetivo é investigar a relação entre esses fatores, analisando  atributos como horas trabalhadas, renda, educação e idade dos membros da família. Além disso, também permite explorar correlações entre  características locais, taxa de desemprego e aspectos do trabalho e da família. Em resumo, o conjunto de dados proporciona uma base para estudos  sobre emprego, família e renda.

## Apresentação das Colunas

* **work:** Indica se a pessoa trabalha ou não (“yes” para trabalha, “no” caso contrário).
* **hoursw:** Número de horas trabalhadas pela eposa por semana.
* **child6:** Número de filhos com idade inferior a 6 anos.
* **child618:** Número de filhos com idade entre 6 e 18 anos.
* **agew:** Idade da esposa.
* **educw:** Anos de educação da esposa.
* **hearnw:** Ganho médio por hora da esposa.
* **wagew:** Salário semanal da esposa.
* **hoursh:** Número de horas trabalhadas pelo marido.
* **ageh:** Idade do marido.
* **educh:** Anos de educação do marido.
* **wageh:** Salário semanal do marido.
* **income:** Renda familiar total.
* **educwm:** Anos de educação da mãe da eposa.
* **educwf:** Anos de educação do pai da esposa.
* **unemprate:** Taxa de desemprego da cidade onde vivem.
* **city:** Indica se a pessoa vive em uma cidade ou não (“yes” para vive em uma cidade, “no” caso contrário).
* **experience:** Anos de experiência da esposa no mercado de trabalho.

## Tratamento de dados

**Horas trabalhadas pela eposa por semana**

```r
media_hoursw<- mean(data$hoursw)
mediana_hoursw<- median(data$hoursw)
moda_hoursw<- mode(data$hoursw)
desviopadrao_hoursw<- sd(data$hoursw)
variancia_hoursw<- var(data$hoursw)
amplimax_hoursw<- max(data$hoursw)
amplimin_hoursw<- min(data$hoursw)
quartis_hoursw<- quantile(data$hoursw)

print(paste0("Média de horas trabalhadas pela eposa por semana: ",  media_hoursw))
print(paste0("Mediana de horas trabalhadas pela eposa por semana: ",  mediana_hoursw))
print(paste0("Moda de horas trabalhadas pela eposa por semana: ",  moda_hoursw))
print(paste0("Devio padrão de horas trabalhadas pela eposa pela eposa por semana: ",  desviopadrao_hoursw))
print(paste0("Variância de horas trabalhadas pela eposa por semana: ",  variancia_hoursw))
print(paste0("Amplitude Máxima de horas trabalhadas pela eposa por semana: ",  amplimax_hoursw))
print(paste0("Amplitude Mínima de horas trabalhadas pela eposa por semana: ",  amplimin_hoursw))
print(paste0("1º Quartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[2]))
print(paste0("3º Quartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[5]))
print(paste0("Interquartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[5] - quartis_hoursw[2]))
```

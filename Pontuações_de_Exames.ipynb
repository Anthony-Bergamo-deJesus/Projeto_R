## Introduo

O conjunto de dados aborda informaes sobre trabalho, famlia e renda. Seu objetivo  investigar a relao entre esses fatores, analisando  atributos como horas trabalhadas, renda, educao e idade dos membros da famlia. Alm disso, tambm permite explorar correlaes entre  caractersticas locais, taxa de desemprego e aspectos do trabalho e da famlia. ## Apresentao das Colunas

* **work:** Indica se a pessoa trabalha ou no (yes para trabalha, no caso contrrio).
* **hoursw:** Nmero de horas trabalhadas pela eposa por semana.
* **child6:** Nmero de filhos com idade inferior a 6 anos.
* **child618:** Nmero de filhos com idade entre 6 e 18 anos.
* **agew:** Idade da esposa.
* **educw:** Anos de educao da esposa.
* **hearnw:** Ganho mdio por hora da esposa.
* **wagew:** Salrio semanal da esposa.
* **hoursh:** Nmero de horas trabalhadas pelo marido.
* **ageh:** Idade do marido.
* **educh:** Anos de educao do marido.
* **wageh:** Salrio semanal do marido.
* **income:** Renda familiar total.
* **educwm:** Anos de educao da me da eposa.
* **educwf:** Anos de educao do pai da esposa.
* **unemprate:** Taxa de desemprego da cidade onde vivem.

## Tratamento de dados

**Horas trabalhadas pela eposa por semana**

```r
media_hoursw<- mean(data$hoursw)
mediana_hoursw<- median(data$hoursw)
moda_hoursw<- mode(data$hoursw)
desviopadrao_hoursw<- sd(data$hoursw)
variancia_hoursw<- var(data$hoursw)
amplimax_hoursw<- max(data$hoursw)
amplimin_hoursw<- min(data$hoursw)
quartis_hoursw<- quantile(data$hoursw)

print(paste0("Mdia de horas trabalhadas pela eposa por semana: ",  media_hoursw))
print(paste0("Mediana de horas trabalhadas pela eposa por semana: ",  mediana_hoursw))
print(paste0("Moda de horas trabalhadas pela eposa por semana: ",  moda_hoursw))
print(paste0("Devio padro de horas trabalhadas pela eposa pela eposa por semana: ",  desviopadrao_hoursw))
print(paste0("Varincia de horas trabalhadas pela eposa por semana: ",  variancia_hoursw))
print(paste0("Amplitude Mxima de horas trabalhadas pela eposa por semana: ",  amplimax_hoursw))
print(paste0("Amplitude Mnima de horas trabalhadas pela eposa por semana: ",  amplimin_hoursw))
print(paste0("1 Quartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[2]))
print(paste0("3 Quartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[5]))
print(paste0("Interquartil de horas trabalhadas pela eposa por semana: ",  quartis_hoursw[5] - quartis_hoursw[2]))
```
